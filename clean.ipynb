```python
import pandas as pd
df = pd.read_csv('Shared_Mobility.csv', delimiter=';')

rows=df.shape[0]
columns=df.shape[1]

#Quantity of data 
print(f"The data has a size of {rows} rows and {columns} columns")
print(df.head(5))
```

```python
df.columns
```

```python
def replace_comma_with_dot(value):
    return str(value).replace(',', '.')

# Apply the replacement function to all columns
df = df.applymap(replace_comma_with_dot)

# Display the updated DataFrame
print(df)
```

```python
import pandas as pd

# Assuming df is your DataFrame
v1 = ['City', 'Country', 'Latitude', 'Longitude', 'Operator/Service provider', 'Mode', 'Operational model', 'Public or private involvement']
map_df = df[v1]

# Save the selected columns to a CSV file
#map_df.to_csv('map_df.csv', index=False)
```

```python
map_df
```

```python
cities=map_df.groupby(['City','Mode'])

for (city,mode), group in cities:  
    print(f"City:{city}, Mode: {mode}")
    print(group)
    print("\n")
```
